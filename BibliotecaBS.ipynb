# [dbo].[MostrarPrestamos]

```sql
/****** Object:  StoredProcedure [dbo].[MostrarPrestamos]    Script Date: 15/05/2024 03:53:49 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--Procedimiento almacenado para mostrar Prestamos
CREATE PROCEDURE [dbo].[MostrarPrestamos]
AS
BEGIN
    SET NOCOUNT ON;

    SELECT P.ID_Prestamo, 
           U.ID_Usuario,
           U.nombreCompleto, 
           L.id_libro, 
           L.tituloLibro,
           P.fecha_prestamo, 
           P.fecha_devolucion
    FROM Prestamo P
    INNER JOIN Usuario U ON P.ID_Usuario = U.ID_Usuario
    INNER JOIN Libro L ON P.ID_Libro = L.id_libro
END
GO
```

# [dbo].[ObtenerTituloLibroYNombreCompleto]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerTituloLibroYNombreCompleto]    Script Date: 15/05/2024 03:53:49 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--Procedimiento almacenado para Obtener Titulo Libro Y Nombre Completo
CREATE PROCEDURE [dbo].[ObtenerTituloLibroYNombreCompleto]
    @ID_Libro INT,
    @ID_Usuario INT
AS
BEGIN
    SET NOCOUNT ON;

    SELECT L.tituloLibro, U.nombreCompleto
    FROM Libro L
    CROSS JOIN Usuario U 
    WHERE L.id_libro = @ID_Libro AND U.ID_Usuario = @ID_Usuario;
END

GO
```
